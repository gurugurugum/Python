```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def f2(x, source):
    return [-1/((x[0]-source[0])**2+(x[1]-source[1])**2)**(3/2)*(x[0]-source[0]), -1/((x[0]-source[0])**2+(x[1]-source[1])**2)**(3/2)*(x[1]-source[1])]

def f3(x, s1, s2):
    return [f2(x,s1)[0]+f2(x,s2)[0], f2(x,s1)[1]+f2(x,s2)[1]]

def x(l, n, id):
    return l[2*n+2*id:2*n+2*id+2]

def v(l, n, id):
    return l[2*id:2*id+2]

def func1(y, t):
    return f2(x(y, 1, 0),[0,0])+[v(y, 1, 0)[0],v(y, 1, 0)[1]]

def func2(y, t):
    return f2(x(y, 2, 0),x(y, 2, 1))+f2(x(y, 2, 1),x(y, 2, 0))+[v(y, 2, 0)[0],v(y, 2, 0)[1]]+[v(y, 2, 1)[0],v(y, 2, 1)[1]]

def func3(y, t):
    return f3(x(y, 3, 0),x(y, 3, 1),x(y, 3, 2))+f3(x(y, 3, 1),x(y, 3, 2),x(y, 3, 0))+f3(x(y, 3, 2),x(y, 3, 0),x(y, 3, 1))+[v(y, 3, 0)[0],v(y, 3, 0)[1]]+[v(y, 3, 1)[0],v(y, 3, 1)[1]]+[v(y, 3, 2)[0],v(y, 3, 2)[1]]
```

```python
x0=[1, 0]
v0=[0, 1]
y0 = v0+x0
t = np.arange(0, 2, 0.01)

traj = odeint(func1, y0, t)

fig = plt.figure()
plt.plot(traj[:, 2], traj[:, 3])
plt.show()
```

```python
x0_0=[0.1, 0]
v0_0=[0, 1]
x1_0=[-0.1, 0]
v1_0=[0, -1]
y0 = v0_0+v1_0+x0_0+x1_0
t = np.arange(0, 0.15, 0.001)

traj = odeint(func2, y0, t)

fig = plt.figure()
plt.plot(traj[:, 4], traj[:, 5])
plt.plot(traj[:, 6], traj[:, 7])
plt.show()
```

```python
x0_0=[0.1, 0]
v0_0=[0, 1]
x1_0=[0, 0.1]
v1_0=[-1, 0]
x2_0=[-0.1, 0]
v2_0=[0, -1]
y0 = v0_0+v1_0+v2_0+x0_0+x1_0+x2_0
t = np.arange(0, 0.3, 0.001)

traj = odeint(func3, y0, t)

fig = plt.figure()
plt.plot(traj[:, 6], traj[:, 7])
plt.plot(traj[:, 8], traj[:, 9])
plt.plot(traj[:, 10], traj[:, 11])
plt.show()
```
